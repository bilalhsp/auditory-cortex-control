```python
# ------------------  set up logging ----------------------
import logging
from auditory_cortex.utils import set_up_logging
set_up_logging('info')

from auditory_cortex.plotters.correlation_plotter import RegPlotter
from auditory_cortex.plotters import tikzplots
%matplotlib inline
```

```python
save_tikz=True
threshold=0.5
lag = 200
```

```python
model_names = [
	'deepspeech2',
	'speech2text',
	'whisper_tiny',
	'whisper_base',
	'wav2vec2',
	'wav2letter_modified',
	]
identifier = f'ucsf_timit_trf_lags{lag}_bw50_ahmed_25'
tikz_indicator = f'trf{lag}'
bin_width = 50
mVocs = False

dist_core = []
dist_non_primary = []
normalize_layer_ids=True
for i, model_name in enumerate(model_names):
    peak_layers, corr_dist = RegPlotter.get_dist_prefered_layer(
            model_name,
            identifier,
            bin_width=bin_width,
            mVocs=mVocs,
            threshold=threshold,
            normalize_layer_ids=normalize_layer_ids,
            )
    dist_core.extend(peak_layers['core'])
    dist_non_primary.extend(peak_layers['non-primary'])

peak_layers = {'core': dist_core, 'non-primary': dist_non_primary}
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde
import numpy as np
fig, ax = plt.subplots(figsize=(6,4))

data = {
    'Core': dist_core,
    'Non-Primary': dist_non_primary,
}
colors = {
    'Core': 'red',
    'Non-Primary': 'tab:blue',
}

for label, d in data.items():
    d = np.array(d)
    ax.hist(d, bins=10, alpha=0.3, label=label, color=colors[label], density=False)
    kde = gaussian_kde(d)
    x = np.linspace(d.min(), d.max(), 300)
    ax.plot(x, kde(x), color=colors[label], linewidth=2)

ax.set_xlabel('Peak Layer', fontsize=14)
ax.tick_params(labelsize=12)
ax.legend(fontsize=12)
ax.set_yticks([])
ax.spines['left'].set_visible(False)
plt.tight_layout()
```
